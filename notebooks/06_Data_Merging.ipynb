```python
import pandas as pd
```

```python
orders = pd.read_csv("../data/raw/olist_orders_dataset.csv")
```

```python
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
```

```python
products = pd.read_csv("../data/raw/olist_products_dataset.csv")
```

```python
orders.shape
```

```python
order_items.shape
```

```python
products.shape
```

```python
orders_items = pd.merge(
    orders,
    order_items,
    on="order_id",
    how="inner"
)
```

```python
orders_items.head()
```

```python
orders_items.shape
```

```python
sales = pd.merge(
    orders_items,
    products,
    on="product_id",
    how="inner"
)
```

```python
sales.head()
```

```python
sales["revenue"] = sales["price"] + sales["freight_value"]
```

```python
sales[["price", "freight_value", "revenue"]].head()
```

```python
category_revenue = sales.groupby("product_category_name")["revenue"].sum()
```

```python
category_revenue = category_revenue.sort_values(ascending=False)
```

```python
top10_revenue = category_revenue.head(10)
```

```python
top10_revenue
```

```python
import matplotlib.pyplot as plt

top10_revenue.plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Top 10 Categories by Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python
top_products = sales.groupby("product_id")["order_id"].count()
```

```python
top_products = top_products.sort_values(ascending=False)
```

```python
top_products.head(10)
```

```python
import matplotlib.pyplot as plt

top_products.head(10).plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Top 10 Selling Products")
plt.xlabel("Product ID")
plt.ylabel("Orders Sold")
plt.xticks(rotation=45)

plt.show()
```

```python
top_customers = sales.groupby("customer_unique_id")["order_id"].count()

top_customers = top_customers.sort_values(ascending=False)

top_customers.head(10)
```

```python
print(sales.columns.tolist())
```

```python
customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
```

```python
customers.head()
```

```python
sales = sales.merge(
    customers,
    on="customer_id",
    how="left"
)
```

```python
sales.columns
```

```python
top_customers = sales.groupby("customer_unique_id")["order_id"].count()

top_customers = top_customers.sort_values(ascending=False)

top_customers.head(10)
```

```python
import matplotlib.pyplot as plt

top_customers.head(10).plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Top 10 Customers")
plt.xlabel("Customer Unique ID")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

plt.show()
```

```python
sellers = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
```

```python
sellers.head()
```

```python
sales = sales.merge(
    sellers,
    on="seller_id",
    how="left"
)
```

```python
sales.columns
```

```python
top_sellers = sales.groupby("seller_id")["order_id"].count()

top_sellers = top_sellers.sort_values(ascending=False)

top_sellers.head(10)
```

```python
top_sellers.head(10).plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Top 10 Sellers")
plt.xlabel("Seller ID")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

plt.show()
```

```python
total_revenue = sales["revenue"].sum()

total_revenue
```

```python
total_orders = sales["order_id"].nunique()

total_orders
```

```python
total_customers = sales["customer_unique_id"].nunique()

total_customers
```

```python
total_sellers = sales["seller_id"].nunique()

total_sellers
```

```python
average_order_value = sales["revenue"].mean()

average_order_value
```

```python
print("Total Revenue :", total_revenue)
print("Total Orders :", total_orders)
print("Total Customers :", total_customers)
print("Total Sellers :", total_sellers)
print("Average Order Value :", average_order_value)
```
