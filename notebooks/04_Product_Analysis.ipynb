```python
top_categories = products["product_category_name"].value_counts().head(10)

top_categories
```

```python
import matplotlib.pyplot as plt

top_categories.plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)

plt.show()
```
