```python
state_customers = customers.groupby("customer_state")["customer_unique_id"].nunique()

state_customers
```

```python
import matplotlib.pyplot as plt

state_customers.sort_values(ascending=False).plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Customers by State")
plt.xlabel("State")
plt.ylabel("Number of Customers")

plt.show()
```
