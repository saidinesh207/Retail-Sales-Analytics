```python
import matplotlib.pyplot as plt

rating_counts = reviews["review_score"].value_counts().sort_index()

rating_counts.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Customer Review Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")

plt.show()
```
